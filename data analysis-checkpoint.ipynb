```python
import pandas as pd 
data=pd.read_csv("Training Data.csv")
data.head()
```

```python
features=data.drop("risk_flag",axis=1)
features.head()
```

```python
label=data["risk_flag"].copy()
```

```python
label.head()
```

```python
features["married"]=features["married"].replace(["single"] ,0)
features["married"]=features["married"].replace(["married"] ,1)
features["car_ownership"]=features["car_ownership"].replace(["no"] ,0)
features["car_ownership"]=features["car_ownership"].replace(["yes"] ,1)
features["house_ownership"]=features["house_ownership"].replace(["rented"] ,1)
features["house_ownership"]=features["house_ownership"].replace(["owned"] ,2)
features["house_ownership"]=features["house_ownership"].replace(["norent_noown"] ,0)
```
